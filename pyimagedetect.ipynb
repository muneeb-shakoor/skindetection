```python
## Importing all essential libraries
import numpy as np
import argparse
import cv2
import matplotlib.pyplot as plt
import skimage.exposure
from PIL import Image
import numpy as np
```

```python
#specifying skin color range 
# minRange (for dark skin color)
minRange = np.array([76,87,84],np.uint8)
# maxRange (for fair skin color)
maxRange = np.array([220,244,220],np.uint8)

#loading image from the same directory
image = cv2.imread('pink.jpg')

# change our image bgr to ycr using cvtcolor() method 
YCRimage = cv2.cvtColor(image,cv2.COLOR_BGR2YCR_CB)

# apply min or max range on skin area in our image
skinArea = cv2.inRange(YCRimage,minRange,maxRange)
detectedSkin = cv2.bitwise_and(image, image, mask = skinArea)

cv2.imwrite("detectedImage.png", np.hstack([image,detectedSkin]))
```

```python
#visualizing skin vs non skin (white is skin, black is non skin)
plt.imshow(cv2.cvtColor(skinArea, cv2.COLOR_BGR2RGB))
```

```python
#extracting skin pixels from the detected skin image
skin=np.where((detectedSkin[:,:,0]!=0) & (detectedSkin[:,:,1]!=0) & (detectedSkin[:,:,2]!=0))
```

```python
#replacing all the pixels of skin area with blue color pixel
image[skin]=(176,134,39)
```

```python
#visualizing processed image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
#cv2.imwrite('Blue_outputt.jpg',image)
```
